```python
from wordcloud import WordCloud
from pathlib import Path
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

```python
data_folder = Path('/users/wg889232/Data/')

# Read the whole text.
dream = open(data_folder/ "DreamSpeech.txt").read()
lear = open(data_folder/ 'KingLear.txt').read()
gettysburg = open(data_folder/ 'Gettysburg.txt').read()

# Bring in images for word cloud
crown_mask = np.array(Image.open(data_folder/ 'crown_mask.jpg'))
dream_mask = np.array(Image.open(data_folder/ 'comment.png'))
getty_mask = np.array(Image.open(data_folder/ 'AL_mask.jpg'))
```

```python
dreamcloud = WordCloud(background_color='black',
                      contour_width= 1,
                      contour_color='black',
                      mask=dream_mask).generate(dream)


plt.imshow(dreamcloud, interpolation='bilinear')
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```

```python
learcloud = WordCloud(background_color='black',
                      contour_width= 1,
                      contour_color='black',
                      mask=crown_mask).generate(lear)


plt.imshow(learcloud, interpolation='bilinear')
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```

```python
gettycloud = WordCloud(background_color='black',
                      contour_width= 1,
                      contour_color='black',
                      mask=getty_mask).generate(gettysburg)


plt.imshow(gettycloud, interpolation='bilinear')
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```
